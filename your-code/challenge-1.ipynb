#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
plt.plot(x, x**2, label='x**2')
plt.plot(x, x**x, label='x**x')

plt.xlabel('x as') 
plt.ylabel('y as')

plt.xscale("log")
plt.yscale("log")

plt.title("y=x^2 and y=exp(x)") 
plt.legend()
plt.show() 

```

```python
fig, [x2, xx] = plt.subplots(nrows = 1,ncols=2, sharey = 'none', figsize=(10,10))
# self.bbox_inches = Bbox.from_bounds(0, 0, *figsize)

x2.plot(x,x**2)
x2.set_title('x**2')   
x2.set_xlabel('x label')     
x2.set_ylabel('y label')

xx.plot(x,x**x)
xx.set_title('x**x')
xx.set_xlabel('x label')
xx.set_ylabel('Z label')
xx.set_xscale('log')
xx.set_yscale('log')


plt.show()
```
